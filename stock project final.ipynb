```python
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
# Question 1 - Extracting Tesla Stock Data Using yfinance 
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")
tesla_data.reset_index(inplace=True)
tesla_data.head()
```

```python
# Question 2 - Extracting Tesla Revenue Data Using Webscraping 
url= "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
html_data=requests.get(url).text
soup = BeautifulSoup(html_data,"html5lib")
tesla_revenue= pd.read_html(url, match="Tesla Quarterly Revenue", flavor='bs4')[0]
tesla_revenue=tesla_revenue.rename(columns = {'Tesla Quarterly Revenue(Millions of US $)': 'Date', 'Tesla Quarterly Revenue(Millions of US $).1': 'Revenue'}, inplace = False)
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].str.replace(",","").str.replace("$","")
tesla_revenue.head()
tesla_revenue.dropna(inplace=True)
tesla_revenue.tail()
```

```python
import yfinance as yf
gamestop = yf.Ticker("GME")
gamestop_stock_data = gamestop.history(period="max")
print(gamestop_stock_data.head())
import requests
from bs4 import BeautifulSoup
import pandas as pd
url = 'https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue'
response = requests.get(url)
response.raise_for_status()  
soup = BeautifulSoup(response.text, 'html5lib')
tables = soup.find_all('table')
def is_revenue_table(table):
    headers = [header.text.strip() for header in table.find_all('th')]
    return "Date" in headers and "Revenue" in headers
print("Inspecting table headers to find the correct one...")
for i, table in enumerate(tables):
    headers = [header.text.strip() for header in table.find_all('th')]
    print(f"Table {i} headers: {headers}")
revenue_table = None
for table in tables:
    if is_revenue_table(table):
        revenue_table = table
        break
if revenue_table is None:
    raise ValueError("Could not find the revenue table on the page.")

revenue_data = []
for row in revenue_table.find_all('tr')[1:]:  # Skip the header row
    cols = row.find_all('td')
    if len(cols) > 1:
        revenue_data.append({
            'date': cols[0].text.strip(),
            'revenue': cols[1].text.strip().replace('$', '').replace(',', '')
        })
gamestop_revenue_df = pd.DataFrame(revenue_data)
print(gamestop_revenue_df)
import plotly.graph_objs as go
from plotly.subplots import make_subplots
gamestop_revenue_df['date'] = pd.to_datetime(gamestop_revenue_df['date'])
fig = make_subplots(rows=2, cols=1, subplot_titles=("GameStop Stock Price", "GameStop Revenue"))
fig.add_trace(go.Scatter(x=gamestop_stock_data.index, y=gamestop_stock_data['Close'], name='Stock Price', line=dict(color='blue')), row=1, col=1)
fig.add_trace(go.Bar(x=gamestop_revenue_df['date'], y=gamestop_revenue_df['revenue'], name='Revenue', marker=dict(color='green')), row=2, col=1)
fig.update_layout(height=800, width=1000, title_text="GameStop Stock Price and Revenue Dashboard")
fig.show()
```
